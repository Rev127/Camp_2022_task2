### Question 1

```python
import pandas as pd
import numpy as np
file_name = 'Energy Indicators.xls'
energy = pd.read_excel(file_name, skiprows= 16, nrows=228)
energy = pd.DataFrame(energy, columns=['Unnamed: 2',	'Energy Supply',	'Energy Supply per capita',	'Renewable Electricity Production'])

energy = energy.where(energy != '...')
energy['Energy Supply'] = energy['Energy Supply']*100000

energy['Unnamed: 2'] = energy['Unnamed: 2'].map(lambda x: str(x)[:str(x).find('(')])
energy['Unnamed: 2'] = energy['Unnamed: 2'].map(lambda x: x.rstrip('1234567890'))

energy['Unnamed: 2'] = energy.where(energy['Unnamed: 2'] != 'Republic of Korea', other= 'South Korea')
energy['Unnamed: 2'] = energy.where(energy['Unnamed: 2'] != 'United States of America', other= 'United States')
energy['Unnamed: 2'] = energy.where(energy['Unnamed: 2'] != 'United Kingdom of Great Britain and Northern Ireland', other= 'United Kingdom')
energy['Unnamed: 2'] = energy.where(energy['Unnamed: 2'] != 'China, Hong Kong Special Administrative Region', other= 'Hong Kong')

energy.head()
```

```python
file_name = "world_bank.csv"
GDP = pd.read_csv(file_name, skiprows=4)
GDP = pd.DataFrame(GDP)

GDP['Country Name'] = GDP.where(GDP['Country Name'] != 'Korea, Rep.', other= 'South Korea')
GDP['Country Name'] = GDP.where(GDP['Country Name'] != 'Iran, Islamic Rep.', other= 'Iran')
GDP['Country Name'] = GDP.where(GDP['Country Name'] != 'Hong Kong SAR, China', other= 'Hong Kong')

GDP.head()
```

```python
file_name = 'scimagojr country rank 1996-2021.xls'
ScimEn = pd.read_excel(file_name)
ScimEn = pd.DataFrame(ScimEn)

ScimEn.head()
```

```python
def answer_one():
    energy.rename(columns= {'Unnamed: 2':'Country Name'}, inplace = True)
    energy.rename(columns= {'Energy Supply per capita':'Energy Supply per Capita'}, inplace = True)
    energy.rename(columns= {'Renewable Electricity Production':'% Renewable'}, inplace = True)
    ScimEn.rename(columns= {'Country' : 'Country Name'}, inplace = True)

    temp = pd.merge(energy, GDP, on='Country Name', how='right')
    temp.head()
    NewDataFrame = pd.merge(temp, ScimEn, on='Country Name', how='right')
    NewDataFrame.set_index('Country Name', inplace = True)
    NewDataFrame = pd.DataFrame(NewDataFrame, columns=['Rank', 'Documents', 'Citable documents', 'Citations', 'Self-citations', 'Citations per document', 'H index', 'Energy Supply', 'Energy Supply per Capita', '% Renewable', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015'])
    return NewDataFrame.head(15)


answer_one()

    
```

### Question 2

```python
def answer_two():
    Top15 = answer_one()
    col = ['2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015']
    avgGDP = pd.DataFrame(index= Top15.index)
    avgGDP['avgGDP'] = Top15[col].sum(axis=1) / 10
    avgGDP = avgGDP.sort_values(['avgGDP'], ascending= False)
    return avgGDP

answer_two()
```

### Question 3

```python
def answer_three():
    Top15 = answer_one()
    avgGDP = answer_two()

    temp_index = 0
    for i in avgGDP.index:
        if temp_index == 5:
            temp1 = Top15.loc[i, '2015']
            temp2 = Top15.loc[i, '2006']
            return temp1 - temp2
        else:
            temp_index +=1


answer_three()
```

### Question 4

```python
def answer_four():
    Top15 = answer_one()
    Top15.insert(5, 'Relation', np.NaN)
    max = 0
    for i in Top15.index:
        temp1 = Top15.loc[i, 'Self-citations']
        temp2 = Top15.loc[i, 'Citations']
        Top15.loc[i, 'Relation'] = (temp1 / temp2) * 100
        if ((temp1 / temp2) * 100) > max:
            max = (temp1 / temp2) * 100

    for i in Top15.index:
         if Top15.loc[i, 'Relation'] == max:
             temp = (i, max)
             return temp

answer_four()
```

### Question 5

```python
def answer_five():
    Top15 = answer_one()
    Top15.insert(1, 'Population', np.NaN)

    for i in Top15.index:
        temp1 = Top15.loc[i, 'Energy Supply']
        temp2 = Top15.loc[i, 'Energy Supply per Capita']
        Top15.loc[i, 'Population'] = temp1 / temp2

    Top15 = Top15.sort_values(['Population'], ascending=False)
    #return Top15

    temp_index = 0
    for i in Top15.index:
         if temp_index == 2:
             return i
         else:
             temp_index += 1

answer_five()
```

### Question 6

```python
def answer_six():
    Top15 = answer_one()
    Top15.insert(6, 'Сitable documents per person', np.NaN)

    for i in Top15.index:
        temp1 = Top15.loc[i, 'Energy Supply']
        temp2 = Top15.loc[i, 'Energy Supply per Capita']
        temp3 = Top15.loc[i, 'Citable documents']
        Top15.loc[i, 'Сitable documents per person'] = temp3 / (temp1 / temp2)

    answer = pd.DataFrame(Top15, columns= ['Сitable documents per person', 'Energy Supply per Capita'])
    data = answer.corr()

    return data

answer_six()
```

### Question 7

```python
def answer_seven():
    Top15 = answer_one()
    ContinentDict  = {'China':'Asia', 
                  'United States':'North America', 
                  'Japan':'Asia', 
                  'United Kingdom':'Europe', 
                  'Russian Federation':'Europe', 
                  'Canada':'North America', 
                  'Germany':'Europe', 
                  'India':'Asia',
                  'France':'Europe', 
                  'South Korea':'Asia', 
                  'Italy':'Europe', 
                  'Spain':'Europe', 
                  'Iran':'Asia',
                  'Australia':'Australia', 
                  'Brazil':'South America'}
    

    NewDataFrame = pd.DataFrame(columns=['size', 'sum', 'mean', 'std'], index=['Asia', 'Australia', 'Europe', 'North America', 'South America'])
    NewDataFrame = NewDataFrame.where(NewDataFrame == np.NaN, other= 0)

    for i in NewDataFrame.index:
        for j in Top15.index:
            temp1 = 0
            temp2 = 0
            if i ==  ContinentDict[j]:
                NewDataFrame.loc[i, 'size'] += 1
                if Top15.loc[j, 'Energy Supply'] != np.NaN: 
                    temp1 = Top15.loc[j, 'Energy Supply']
                if Top15.loc[j, 'Energy Supply'] != np.NaN:
                    temp2 = Top15.loc[j, 'Energy Supply per Capita'] 
                NewDataFrame.loc[i, 'sum'] += temp1 / temp2
                NewDataFrame.loc[i, 'mean'] += temp1 / temp2
        NewDataFrame.loc[i, 'mean'] /= NewDataFrame.loc[i, 'size']
    
                

    return NewDataFrame

answer_seven()
```
